```python
def get_original_lables(test_final):
  original_lables = []
  for i in range(0, len(test_final)):
    original_lables.append(test_final[i][5])

  return original_lables
```

```python
def delete_last_column(data):
  return data[:, :-1]  # Select all rows and all columns except the last
```

```python
def split_dataset(X_updated, Y_updated):
  # Split data into training and test sets

  # Further split class 1 for training and testing (40 train, 10 test)
  X_train_C1, X_test_C1, y_train_C1, y_test_C1 = train_test_split(X_updated[Y_updated == 1], Y_updated[Y_updated == 1], test_size=0.2, random_state=42)
  print("1st Class Dimensionalities")

  print("Training samples shape ", X_train_C1.shape)
  print("Testing samples shape ", X_test_C1.shape)
  print("Training lables shape ", y_train_C1.shape)
  print("Testing lables shape ", y_test_C1.shape)

  # Further split class 1 for training and testing (80 train, 20 test)
  X_train_C2, X_test_C2, y_train_C2, y_test_C2 = train_test_split(X_updated[Y_updated == -1], Y_updated[Y_updated == -1], test_size=0.2, random_state=42)

  print("================================")

  print("1st Class Dimensionalities")
  print("Training samples shape ", X_train_C2.shape)
  print("Testing samples shape ", X_test_C2.shape)
  print("Training lables shape ", y_train_C2.shape)
  print("Testing lables shape ", y_test_C2.shape)


  # Combine training data for C1 and C2
  X_train_final = np.concatenate((X_train_C1, X_train_C2))
  y_train_final = np.concatenate((y_train_C1, y_train_C2))


  # Combine testing data for C1 and C2
  X_test_final = np.concatenate((X_test_C1, X_test_C2))
  y_test_final = np.concatenate((y_test_C1, y_test_C2))

  return X_train_final, y_train_final, X_test_final, y_test_final
```

```python
def calculate_best_fit_weight(X_train, y_train):
  best_fit_weight = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
  print(best_fit_weight)
  return best_fit_weight
```

```python
def decide_class(line1, line2, line3, test_sample):
  sample_sign_list =[]

  line1_sign = np.sign(np.dot(test_sample, line1))
  line2_sign = np.sign(np.dot(test_sample, line2))
  line3_sign = np.sign(np.dot(test_sample, line3))

  sample_sign_list.append(line1_sign)
  sample_sign_list.append(line2_sign)
  sample_sign_list.append(line3_sign)

  positive_indices = []
  negative_indices = []
  for i, num in enumerate(sample_sign_list):
    if num > 0:
      positive_indices.append(i)
    if num < 0:
      negative_indices.append(i)

  print(test_sample)
  print(positive_indices)
  print(negative_indices)

  return positive_indices
```

```python
def calculate_accuracy(actual, counter):
  accuracy = counter/ len(actual) * 100

  print("The accuracy of this model is ", accuracy)
```

```python
counter = 0
for i, test_sample in enumerate(X_test_final):
    predicted_class = decide_class(best_fit_weight1, best_fit_weight2, best_fit_weight3, test_sample)
    actual_class_label = y_test_original_lables[i]  # Access the corresponding actual class label

    print("Actual class label:", actual_class_label)
    print("Predicted class label:", predicted_class)
    if actual_class_label in predicted_class:
      counter += 1


    print("================================")

print(X_test_final.shape)
print(y.shape)

calculate_accuracy(y_test_original_lables, counter)
```
